```python
class DisjointSets:
    def __init__(self) -> None:
        self.parent = {}
        self.rank = {}
        self.count = 0

    def __len__(self) -> int:
        return self.count

    def makeSets(self, universe):
        for i in universe:
            if i not in self.rank:
                self.parent[i] = i
                self.rank[i] = 0
                self.count += 1

    def find(self, k):
        if self.parent[k] != k:
            self.parent[k] = self.find(self.parent[k])
        return self.parent[k]

    def union(self, a, b):
        x = self.find(a)
        y = self.find(b)

        if x != y:
            self.count -= 1
            if self.rank[x] > self.rank[y]:
                self.parent[y] = x
            elif self.rank[x] < self.rank[y]:
                self.parent[x] = y
            else:
                self.parent[x] = y
                self.rank[y] += 1


from itertools import product
from typing import List

class Solution:
    def numIslands(self, grid: List[List[str]]) -> int:
        ds = DisjointSets()
        m = len(grid)
        n = len(grid[0])
        ds.makeSets((i, j) for i, j in product(
            range(m), range(n)) if grid[i][j] == "1")

        for i in range(m):
            for j in range(n):
                if grid[i][j] == "1":
                    if i != m - 1 and grid[i + 1][j] == "1":
                        ds.union((i, j), (i + 1, j))
                    if j != n - 1 and grid[i][j + 1] == "1":
                        ds.union((i, j), (i, j + 1))
        return len(ds)
```
